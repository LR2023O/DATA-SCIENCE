```python
import matplotlib.pyplot as plt
import numpy as np

# Data
months = ['07/2019', '08/2019', '09/2019', '10/2019', '11/2019']
searches = [50, 53, 59, 56, 62]
direct = [39, 47, 42, 51, 51]
social_media = [70, 80, 90, 87, 92]

# Bar width and positions
x = np.arange(len(months))
bar_width = 0.25

# Plot bars
plt.bar(x - bar_width, searches, width=bar_width, color='skyblue', label='Searches')
plt.bar(x, direct, width=bar_width, color='lightcoral', label='Direct')
plt.bar(x + bar_width, social_media, width=bar_width, color='orange', label='Social Media')

for i, v in enumerate(searches):
    plt.text(x[i] - bar_width, v + 1, str(v), ha='center', va='bottom')
for i, v in enumerate(direct):
    plt.text(x[i], v + 1, str(v), ha='center', va='bottom')
for i, v in enumerate(social_media):
    plt.text(x[i] + bar_width, v + 1, str(v), ha='center', va='bottom')

#labels and titles
plt.xlabel("months")
plt.ylabel("visitors in thousands",rotation=0, labelpad=50)
plt.title("Visitors by web traffic sources")

plt.annotate('', xy=(1.1, 0), xytext=(0, 0),
             xycoords=('axes fraction', 'axes fraction'),
             arrowprops=dict(arrowstyle="->", lw=1.0))

plt.annotate('', xy=(0, 1.1), xytext=(0, 0),
             xycoords=('axes fraction', 'axes fraction'),
             arrowprops=dict(arrowstyle="->", lw=1.0))
plt.xticks(x, months)
plt.box(False)
plt.yticks(np.arange(0, 101, 20))
plt.legend(loc='upper center',bbox_to_anchor=(0.4, -0.15),ncol=3,frameon=False)
plt.show()
```
